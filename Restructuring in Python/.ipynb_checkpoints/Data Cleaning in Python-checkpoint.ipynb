```python
import pandas as pd

# Outlining my data transformation, since I am less confident in Python compared to R I feel far more comfortable changing each piece bit by bit rather than doing what should be a substantial data clean into a single code chunk. However in future that will be a priority on the skillset. 
# One of the helpful aspects of this project is that the raw data is already really well categorised. However I think I may come back to this part and add a few more columns when inspiration strikes. 
# Potentially I am already thinking about categorising the size of loan but that can wait for now. 
```

```python
loans_raw = pd.read_csv("../raw data/lending_club_loans.csv")

# Downloading the raw data. 
```

```python
loans_raw.head()

# taking a sneak peak at the data, I did also split screen this and have the raw csv file up to keep a closer eye on it. 
```

```python
loans_clean = loans_raw.drop(loans_raw.columns[53:114], axis = 1)

#as much as I am not a fan of getting rid of a swathe of columns, in this case it was required especially since so many of these were filled with NAs.
```

```python
loans_clean = loans_clean[["loan_amnt", "funded_amnt", "funded_amnt_inv", "term", "int_rate", "installment", "sub_grade", "emp_title", "emp_length", "home_ownership", "annual_inc", "verification_status", "issue_d", "loan_status", "desc", "purpose", "title", "addr_state", "dti", "fico_range_low", "fico_range_high", "total_pymnt", "total_pymnt_inv", "total_rec_prncp", "total_rec_prncp", "total_rec_int", "last_pymnt_d", "last_pymnt_amnt", "last_credit_pull_d", "last_fico_range_high", "last_fico_range_low"]]

# Taking the other half of the columns and taking the priority ones.
```

```python
loans_clean.head()
```

```python
loans_clean.isna().sum()

# finding my nas
```

```python
loans_clean = loans_clean.dropna(subset=["loan_amnt"])

#thankfully there were only 3 but they got out pretty easily. 
```

```python
loans_clean['loan_status'].value_counts()

# Initially this was a wide range but for a logistic regression model it needs to be cut down to a binary option of paid and unpaid.
```

```python
# Create an independent copy of loans_clean to avoid SettingWithCopyWarning
loans_clean = loans_clean.copy()

# Perform the replacement operation
loans_clean['loan_status'] = loans_clean['loan_status'].replace({
    "Does not meet the credit policy. Status:Fully Paid": "Fully Paid",
    "Late (31-120 days)": "Fully Paid",
    "Late (16-30 days)": "Fully Paid",
    "Does not meet the credit policy. Status:Charged Off": "Charged Off"
})

# This may have been an error in my coding but it worked better when I made a copy. Then the next stage was recategorising the defaulted and fully paid loans even though they did not meet credit policy.
# I took out Current, Grace Period and Default because I only wanted the status of finalised loans.
```

```python
loans_clean = loans_clean[~loans_clean['loan_status'].isin(["Current", "In Grace Period", "Default"])]
# Neither decision mattered due to their tiny number but I decided to keep the late payments because the objective is to analyse the default rate and how to reduce it so being up to 4 months late does not matter. 
```

```python
loans_clean.head()
```

```python
state_names = pd.read_csv("../raw data/state_names_info.csv")
grades = pd.read_csv("../raw data/grade_info.csv")

#downloading raw datasets for the joins, or merges in Python. 
```

```python
loans_clean = pd.merge(loans_clean, state_names, left_on='addr_state', right_on='state_abb')
loans_clean = pd.merge(loans_clean, grades, on='sub_grade')

#done my merges but there was a mistake here as I had multiple state_abbs. 
```

```python
loans_clean['emp_title'] = loans_clean['emp_title'].str.title()
loans_clean['home_ownership'] = loans_clean['home_ownership'].str.title()
loans_clean['title'] = loans_clean['title'].str.title()
loans_clean['purpose'] = loans_clean['purpose'].str.replace('_', ' ').str.title()
loans_clean['total_pymnt'] = loans_clean['total_pymnt'].round(2)

# Here I did not like the formatting of the columns so I changed them all to be case title and purpose to lose the lower case. 
# the total_pymnt column also looked messy based on the huge number of digits below decimal point. 
```

```python
pd.set_option('display.max_columns', None)

loans_clean.head()

#Found a way to look at all columns when using head function which I think is far more useful as I want to check all columns are clean. Will be using this more often.
```

```python
loans_clean["issue_d"] = pd.to_datetime(loans_clean["issue_d"], format = "%b-%Y")
loans_clean["last_pymnt_d"] = pd.to_datetime(loans_clean["last_pymnt_d"], format = "%b-%Y")
loans_clean["last_credit_pull_d"] = pd.to_datetime(loans_clean["last_credit_pull_d"], format = "%b-%Y")

# When creating visualisations I know I want to use the dates which I did not do in my final project in R as I had very limited time and knew it was going to be significant to my presentation so instead took a shortcut. This time it's not so needed. 
```

```python
loans_clean = loans_clean.drop(columns=['state_abb_x', 'state_abb_y', 'state_name_y', 'last_fico_range_low', 'total_pymnt', 'fico_range_low', 'funded_amnt'])

# One last thing is to take out all the superfluous columns that seemed useful but after thoroughly checking are essentially duplicate information.
```

```python
loans_clean.to_csv('loans_clean.csv', index=False)

# Saved as a new and clean csv file. 
```
